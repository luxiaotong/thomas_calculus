```python
import numpy as np
from sklearn.linear_model import LinearRegression
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 55
x = np.array([0.20, 0.65, 1.13, 2.55, 4.00, 5.75, 7.80, 10.20, 12.90, 16.00, 18.40])
y = np.array([1.8, 3.6, 5.4, 7.2, 9.0, 10.8, 12.6, 14.4, 16.2, 18.0, 19.8])
plt.plot(x, y, 'o')

def func(x, a, b):
    return a * (x ** b)

popt, pcov = curve_fit(func, x, y)
print(popt)
pred = func(x, popt[0], popt[1])
plt.plot(x, pred)

pred_11 = func(11, popt[0], popt[1])
plt.plot(11, pred_11, 'ro')

X = x.reshape(-1, 1)
reg = LinearRegression().fit(X, y)
pred = x * reg.coef_ + reg.intercept_
plt.plot(x, pred)
plt.plot(11, 11 * reg.coef_ + reg.intercept_, 'yo')
```

```python
# 56
## Least squares polynomial fit.
x = np.array([20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80])
y = np.array([42, 56, 73.5, 91.5, 116, 142.5, 173, 209.5, 248, 292.5, 343, 401, 464])
z = np.polyfit(x, y, 2)
pred = z[0] * (x ** 2) + z[1] * x + z[2]

pred_72 = z[0] * (72 ** 2) + z[1] * 72 + z[2]
print("72: ", pred_72)
pred_85 = z[0] * (85 ** 2) + z[1] * 85 + z[2]
print("85: ", pred_85)

plt.plot(x, y, 'o')
plt.plot(x, pred)
plt.plot(72, pred_72, 'o')
plt.plot(85, pred_85, 'o')

## Linear Regression
X = x.reshape(-1, 1)
reg = LinearRegression().fit(X, y)
pred = x * reg.coef_ + reg.intercept_
plt.plot(x, pred)
plt.plot(72, 72 * reg.coef_ + reg.intercept_, 'o')
plt.plot(85, 85 * reg.coef_ + reg.intercept_, 'o')
```
