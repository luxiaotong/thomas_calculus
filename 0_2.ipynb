```python
import numpy as np
from sklearn.linear_model import LinearRegression
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 55
x = np.array([0.20, 0.65, 1.13, 2.55, 4.00, 5.75, 7.80, 10.20, 12.90, 16.00, 18.40])
y = np.array([1.8, 3.6, 5.4, 7.2, 9.0, 10.8, 12.6, 14.4, 16.2, 18.0, 19.8])
plt.plot(x, y, 'o')

def func(x, a, b):
    return a * (x ** b)

popt, pcov = curve_fit(func, x, y)
print(popt)
pred = func(x, popt[0], popt[1])
plt.plot(x, pred)

print(func(11, popt[0], popt[1]))
```
