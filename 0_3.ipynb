```python
import numpy as np
from sklearn.linear_model import LinearRegression
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 41
x = np.array([50, 60, 70, 80, 90])
y = np.array([25.8, 34.9, 48.2, 66.8, 81.1])
ax = plt.subplot()
plt.plot(x, y, 'o')
plt.xticks(np.arange(0,110,10), np.arange(1900,2010,10))
ax.set_xlim(0, 100)
plt.grid(True)

def func(x, a, b):
    return a * np.exp(x*b)
popt, pcov = curve_fit(func, x, y,  p0=(4, 0.1))
print(popt)

pred_x = np.arange(0, 100, 10)
pred_y = func(pred_x, popt[0], popt[1])
print("1900:", pred_y[0])
plt.plot(pred_x, pred_y)
plt.plot(0, func(0, popt[0], popt[1]), 'o')
```
